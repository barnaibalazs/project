# A vörös és a fehér borok közötti különbség dobozábrán 'density' (sűrűség) alapján, és
# A vörös és a fehér borok közötti különbség hegedű diagramon 'total sulfur dioxide' (összes kén-dioxid) alapján


```python
red=df[df['type']=='red']
white=df[df['type']=='white']
fig,ax=plt.subplots(1,2,figsize=(10,5))
sns.boxplot(data=red,y='density',ax=ax[0])
ax[0].set_title('Vörös')
sns.boxplot(data=white,y='density',ax=ax[1])
ax[1].set_title('Fehér')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
sns.violinplot(data=red,y='total sulfur dioxide',ax=ax[0])
ax[0].set_title('Vörös')
sns.violinplot(data=white,y='total sulfur dioxide',ax=ax[1])
ax[1].set_title('Fehér')
plt.show()
```

# Oszlopok közötti kapcsolatok vizsgálata korrelációs mátrixuk felhasználásával, melyeknek értékeit hőtérképen jeleníti meg

```python
fig = plt.figure(figsize=(10,7))
num_df=df.select_dtypes(include='number')
corr=num_df.corr()
sns.heatmap(corr,annot=True,cmap='coolwarm',fmt=".2f")
plt.title('Hőtérkép')
plt.show
```

# Az adathalmazban szereplő 'fixed acidity', 'density', 'total sulfur dioxide', 'alcohol', 'quality' és 'pH' oszlopok közötti kapcsolatok megjelenítése pairplot alkalmazásával.
# Bortípusok szerinti színezés a 'type' oszlop adatai alapján.  

```python
cols = ['fixed acidity', 'density', 'total sulfur dioxide', 'alcohol', 'quality', 'pH']
sns.pairplot(df[cols + ['type']], hue='type', diag_kind='kde')
plt.suptitle('Pairplot', y=1.02)
plt.show()
```

# Az 'alcohol' és 'density' oszlopok közötti kapcsolat vizsgálata lineplot alkalmazásával.
# Ugyanez a kapcsolatot csak a jobb minőségű borokra történő vizsgálat, ahol a 'quality' oszlopban szereplő érték legalább 7.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.lineplot(data=df, x='alcohol', y='density')
plt.title('Az alcohol és a density közötti kapcsolat')
plt.xlabel('Alcohol')
plt.ylabel('Density')
plt.show()
```

```python
df_good = df[df['quality'] >= 7]
sns.lineplot(data=df_good, x='alcohol', y='density')
plt.title('Alcohol–Density kapcsolat')
plt.xlabel('Alcohol')
plt.ylabel('Density')
plt.show()
```
